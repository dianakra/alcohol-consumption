## This information contains the data averages for each country. We will use these averages to perform a linear regression 

```python
import pandas as pd
import csv
import numpy as np
```

```python
def mean_data (df,str):
    df[str]=df.loc[:, df.columns != 'Unnamed: 0.1'].apply(np.mean,axis=1)
    df1=df[['Unnamed: 0.1',str]]
    return df1
```

## first: Beer drinkers average around the world by contries

```python
BEER_DATA = pd.read_csv(r"data\Clean_Beer_data.csv")
str='beer mean'
MEAN_BEER=mean_data(BEER_DATA,str)
MEAN_BEER.to_csv(r'data\MEAN_BEER.csv')
MEAN_BEER
```

# SECOND: Wine drinkers average around the world by contries

```python
Wine_DATA = pd.read_csv(r"data\Clean_WINE_data.csv")
str='Wine mean'
MEAN_Wine=mean_data(Wine_DATA,str)
MEAN_Wine.to_csv(r'data\MEAN_WINE.csv')
MEAN_Wine
```
