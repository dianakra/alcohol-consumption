# cleaning the data

cleaning the data
first of all we fill the data with NaN where there was no data or data that we dont need, after this we will make the data type to float(in the original it was an object type data) and then delete the rows that there is to many missing data. becouse the 4 data:beer consumbtion,wine consumption,spirits consumption and alcohol consumption have the same structure we will make a function that does all this to the four datasets.

```python
def cleaning_data_format_1(df):
    df=df.loc[:, df.columns != 'Unnamed: 0']
    df=df.loc[:, df.columns != 'Latest dataavailable']
    for column in df.iloc[:,1:]:
        df[column] = df[column].replace(['-'],np.NaN)
    for column in df.iloc[:,1:]:
        df[column] = df[column].astype(float) 
    #df=df.rename({'Unnamed: 0.1':'country'},axis='columns')
    return df
```

## first data is: Beer

```python
Beer_df = pd.read_csv(r"data\list_of_BEER_consumption.csv")
Beer_df.head(4)
```

```python
Clean_Beer_data=cleaning_data_format_1(Beer_df)
```

```python
deaths_by_cause = pd.read_csv(r"data\annual-number-of-deaths-by-cause (1).csv")
```

```python
def cleaning_data_format_2(df):
    df1=df[['Entity','Year','Deaths - Alcohol use disorders - Sex: Both - Age: All Ages (Number)','Deaths - Cirrhosis and other chronic liver diseases - Sex: Both - Age: All Ages (Number)']]
    df1=df1.rename({'Entity':'country'},axis='columns')
    for column in df1.iloc[:,2:]:
        df1[column] = df1[column].astype(float)
    return df1
```

```python
CLEAN_eaths_becouse_alcohol_DATA=cleaning_data_format_2(deaths_by_cause)
```
